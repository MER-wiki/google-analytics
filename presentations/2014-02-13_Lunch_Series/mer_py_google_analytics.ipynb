## Where do users spend their time?

### Time per course

```python
labels = []
total_time = np.zeros(len(list_courses(dict_all)))
for num,course in enumerate(list_courses(dict_all)):
    labels.append(course)
    try:
        total_time[num] = dict_all[course]['visit_duration']
    except:
        pass
    for exam in list_exams(dict_all,course):
        try:
            total_time[num] += dict_all[course][exam]['visit_duration']
        except:
            pass
        for question in list_questions(dict_all,course,exam):
            total_time[num] += dict_all[course][exam][question]['visit_duration']            
plt.figure(figsize(8,8))
plt.pie(total_time, labels = labels, shadow = True, autopct = '%.1f%%', )
plt.title('Total time spent on all wiki pages by course')
plt.savefig('total_time_per_course.png')      
plt.show()
```
