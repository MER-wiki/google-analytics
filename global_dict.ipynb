```python
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
from MERhelpers import *
```

Load data sets, just for the Fall and Winter periods

```python
dict_F12,clicks_F12 = data_to_info_clicks('Analytics_Fall2012.csv')
dict_W13,clicks_W13 = data_to_info_clicks('Analytics_Winter2013.csv')
dict_F13,clicks_F13 = data_to_info_clicks('Analytics_Fall2013.csv')
```

## Total number of clicks

```python
plt.figsize(5,4)
plot_clicks(clicks_F12)
plot_clicks(clicks_W13)
plot_clicks(clicks_F13)
plt.legend(['Fall 2012','Winter 2013','Fall 2013'],'upper left')
plt.title('Total number of clicks: ' + str(sum(clicks_F12) + sum(clicks_W13) + sum(clicks_F13)))
plt.savefig('total_number_of_clicks.png')
```

## Where do users spend their time?

### Time per page type

```python
avg_times = fraction_course_exam_question_by_key(dict_F13,'avg_time')
plt.figsize(6,6)
plt.pie(avg_times, labels=['Course','Exam','Question'], shadow=True, autopct='%.1f%%', explode=(0.1, 0, 0.1))
plt.title('Average time on page per page type')
plt.savefig('avg_time_per_page_type.png')
```

### Clicks per exam

```python
plt.figsize(13,8)
plt.subplots_adjust(hspace=0.6)
for count,course in enumerate(list_courses(dict_W13)):
    plt.subplot(3,2,count+1)
    plot_visit_duration_dist(dict_W13, course, 'bar')
plt.savefig('total_pageview_per_exam.png')
```

### Time per question

```python
plt.figsize(13,8)
plt.subplots_adjust(hspace=0.6)
count = 0
for course in list_courses(dict_W13):
    for exam in list_exams(dict_W13, course):
        if len(list_questions(dict_W13, course, exam)) > 1:
            count += 1
        plt.subplot(3,2,count)
        plot_question_dist(dict_W13, course, exam, 'avg_time', 'bar')
plt.savefig('avg_time_per_question.png')
```

```python
plt.figsize(13,13)
plt.subplots_adjust(hspace=0.6)
count = 0
for course in list_courses(dict_W13):
    for exam in list_exams(dict_W13, course):
        if len(list_questions(dict_W13, course, exam)) > 1:
            count += 1
        plt.subplot(3,2,count)
        plot_question_dist(dict_W13, course, exam, 'pageviews', 'pie')
plt.savefig('pageviews_per_question.png')
```

```python
plt.figsize(9,9)
avg_exam_time = get_avg_exam_time(dict_W13)
plt.subplots_adjust(hspace=0.4)
for count,course in enumerate(avg_exam_time):
    plt.subplot(2,2,count)    
    print_avg_exam_time(avg_exam_time)
plt.savefig('avg_exam_time.png')
```
